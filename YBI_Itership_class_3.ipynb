# Simple Linear Regression

```python
# Step1: import library
import pandas as pd
```

```python
# step2: import data
salary=pd.read_csv("https://github.com/YBI-Foundation/Dataset/raw/main/Salary%20Data.csv")
```

```python
salary.head()
```

```python
salary.info()
```

```python
salary.isnull().sum()
```

```python
salary.describe()
```

```python
salary.columns
```

```python
salary.shape
```

```python
# step3: define target (y) and features (x)
# y - dependent  and x- independent
```

```python
salary.columns
```

```python
Y=salary['Salary']
X=salary[['Experience Years']]
```

```python
salary.shape
```

```python
X.shape
```

```python
Y.shape
```

```python
 # step4: train test split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train, Y_test=train_test_split(X,Y,train_size=0.7,random_state=2529)
X_train
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
# step5: Select Model
from sklearn.linear_model import LinearRegression
model=LinearRegression()   # gives best fit line and equation of line is Y=mX + c
```

```python
# step6 : train model(fit model)
model.fit(X_train,Y_train)
```

```python
model.intercept_  # it gives  (c)
```

```python
model.coef_  # it gives slope (m)
```

```python
# step7: model prediction
Y_pred=model.predict(X_test)
```

```python
# step8: accuracy
from sklearn.metrics import mean_absolute_percentage_error


```

```python
mean_absolute_percentage_error(Y_test, Y_pred)
```

# Multiple Linear Regression:

```python
# step1 : import library
import pandas as pd
```

```python
# step2 : import data
house=pd.read_csv("https://github.com/YBI-Foundation/Dataset/raw/main/Boston.csv")
```

```python
house.head()
```

```python
house.shape
```

```python
house.info()
```

```python
house.columns
```

```python
# step3: define target(Y) and feature(X)
Y=house['MEDV']
X=house[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
```

```python
#step4: train test split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,Y_train, Y_test=train_test_split(X,Y,train_size=0.7,random_state=2529)
```

```python
X_train
```

```python
# check the shape of train and test sample
X_train.shape, X_test.shape,Y_train.shape, Y_test.shape
```

```python
# step5: select model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
# step6: train and fit model
model.fit(X_train, Y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
#step7: predict the model
Y_pred=model.predict(X_test)
```

```python
Y_pred
```

```python
#step8: model accuracy
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(Y_test,Y_pred)
```
