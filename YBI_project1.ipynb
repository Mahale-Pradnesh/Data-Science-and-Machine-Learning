# **Chances of Admission for Higher Studies:**
predict the chances of admission of a student to a Graduate program based on :


1.   GRE Scores (290 to 340)
2.   TOEFL Scores (92 to 120)
3.   University Rating (1 to 5)
4.   Statement of Purpose (1 to 5)
5.   Letter of Recommandation Strength (1 to 5)
6.   Undergraduate CGPA (6.8 to 9.92)
7.   Research Experience (0 to 1)
8.   Chances of Admit (0.34 to 0.97)



```python
# step1: import library
import pandas as pd
```

```python
# step2: import data
admission=pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Admission%20Chance.csv')
```

```python
# Shows the first 5 rows of the dataset.
admission.head()
```

```python
# Gives the information of the dataset
admission.info()
```

```python
# It shows mean, median, mode of the data
admission.describe()
```

```python
# Shows the columns names in the datasets
admission.columns
```

```python
# step3: define target (y) and features(x)
Y=admission['Chance of Admit ']
X=admission[['Serial No', 'GRE Score', 'TOEFL Score', 'University Rating', ' SOP',
       'LOR ', 'CGPA', 'Research']]
```

```python
# step4: train test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.7,random_state=2529)
```

```python
X_train
```

```python
# check shape of train test split sample
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
#step5: select model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
# step6: train or fit model
model.fit(X_train,Y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# step7:predict model
Y_pred=model.predict(X_test)
```

```python
Y_pred
```

```python
#step8 : model accuracy
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(Y_test, Y_pred)
```
