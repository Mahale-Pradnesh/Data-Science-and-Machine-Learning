```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

#Step 1: Load the dataset (Iris dataset as an example)
data = load_iris()
x=data.data
y=data.target

#step 2: Standardize the data
scaler = StandardScaler()
X_std = scaler.fit_transform(x)

#compute the covarience matrix
cov_matrix= np.cov(X_std, rowvar=False)

#Compute the Eigenvectors and Eigenvalues
eigenvalues, eigenvectors =np.linalg.eig(cov_matrix)

#step 5:Calculate the Explained Varience Ratios
explained_variance_ratio = eigenvalues / sum(eigenvalues)

#step 6: Determine the number of principle components
cumulative_variance = np.cumsum(explained_variance_ratio)
num_components= np.argmax(cumulative_variance >= 0.95)+1    #keep components explaining at least 95% variance

#step 7: Project data on to new space
pca= PCA(n_components=num_components)
X_pca=pca.fit_transform(X_std)

#step 8: Visulize the results
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0],X_pca[:,1],c=y, cmap=plt.cm.Set1, edgecolor="k")
plt.title('PCA of Iris Dataset')
plt.xlabel(f'Principle component 1(EVR={explained_variance_ratio[0]:.2f})')
plt.ylabel(f'Principle component 2(EVR={explained_variance_ratio[1]:.2f})')
plt.show()

```
